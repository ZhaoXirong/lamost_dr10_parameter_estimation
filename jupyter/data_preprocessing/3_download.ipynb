```python
import wget
import os
```

```python
# 创建文件夹
fold_path = "./FITSDATA"

for i in range(5,10):
    fold_name = "{}_{}".format(i*10,(i+1)*10)
    path = os.path.join(fold_path,fold_name)
    os.makedirs(path, exist_ok=True)
    print(path)
    
path = os.path.join(fold_path,"100_999")
os.makedirs(path, exist_ok=True)
print(path)
```

```python
for i in range(5,10):
    url_txt_path = os.path.join(fold_path,"url-{}_{}.txt".format(i*10,(i+1)*10))
    download_path = os.path.join(fold_path,"{}_{}".format(i*10,(i+1)*10))
    print(url_txt_path)
    print(download_path)
    f = open(url_txt_path, 'r')
                                

    ls = []
    f.readline()
    while True:
        line = f.readline()
        if line:
            ls.append(line)
        else:
            break

    # Number of all spectra
    print("Number of spectra to download：", len(ls))
    
    for j in range(0, len(ls)):
        try:
            pass
            wget.download(ls[j], download_path)
        except: 
            print(i*10,(i+1)*10,":{}  err".format(j))
            continue
            
url_txt_path = os.path.join(fold_path,"url-100_999.txt")
download_path = os.path.join(fold_path,"100_999".format(i*10,(i+1)*10))

print(url_txt_path)
print(download_path)
f = open(url_txt_path, 'r')

ls = []
f.readline()
while True:
    line = f.readline()
    if line:
        ls.append(line)
    else:
        break

# Number of all spectra
print("Number of spectra to download：", len(ls))

for j in range(0, len(ls)):
    try:
        pass
        wget.download(ls[j], download_path)
    except: 
        print("100_999 :{}  err".format(j))
        continue
```
